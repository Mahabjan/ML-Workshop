1. Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•
2
for ùë• values ranging from -10 to 10.
o Label the x-axis as "X Values" and the y-axis as "Squared Values."
o Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Plot of y = x^2")
plt.show()
```

2. Bar Plot
o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
o Add labels to the x-axis and y-axis and give the chart a title.

```python
import matplotlib.pyplot as plt
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
plt.bar(students, marks, color='skyblue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Scored by Students')
plt.show()
```

3. Scatter Plot
o Generate two lists of 20 random integers each (for x and y coordinates).
o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.

```python
import pandas as pd
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
df = pd.DataFrame({'x': x, 'y': y})
plt.scatter(df['x'], df['y'], c='pink', s=100, edgecolors='black')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Random Scatter Plot with Customizations')
plt.show()
```

4. Histogram
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
o Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=15, color='violet', edgecolor='black')
plt.grid(True)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Simulated Normal Data')
plt.show()
```

5. Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
o Display the percentage values on the chart and use a color palette for
differentiation.

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['lightblue', 'lightgreen', 'violet', 'lightskyblue', 'pink']
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Percentage Distribution of Categories')
plt.show()
```

6. Box Plot (Seaborn)
o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
o Highlight the median and any outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
# Creating a box plot
sns.boxplot(data=marks, color='pink')
# Adding title and labels
plt.title('Box Plot of Student Marks')
plt.ylabel('Marks')
# Displaying
plt.show()
```

7. Heatmap (Seaborn)
o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.
o Add a color bar and annotate the heatmap with the values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# Generate a 5x5 matrix with random integers between 1 and 10
data = np.random.randint(1, 11, size=(5, 5))
# Creating a heatmap
sns.heatmap(data, annot=True, cmap='YlGnBu', cbar=True, fmt='d')
# Adding title
plt.title('Heatmap with Random Integers')
# Displaying
plt.show()
```

8. Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
data = {'Category': ['A', 'B', 'C', 'D', 'E'], 'Values': [5, 7, 3, 8, 4]}
df = pd.DataFrame(data)

# Matplotlib Bar Chart
plt.bar(df['Category'], df['Values'], color='purple', edgecolor='black')
plt.title('Matplotlib Barchart', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.show()

# Seaborn Bar Chart
sns.barplot(data=df, x='Category', y='Values', hue='Category', dodge=False, palette='Blues_d', legend=False)
plt.title('Seaborn Barchart', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.show()
```

9. Pair Plot (Seaborn)
o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot

```python
data = {
    'Age': [25, 30, 35, 40, 28, 33, 38, 29, 31, 36],
    'Height': [160, 165, 170, 175, 162, 168, 172, 164, 167, 169],
    'Weight': [55, 65, 70, 75, 60, 68, 72, 63, 66, 69]
}
df = pd.DataFrame(data)

# Create a pair plot without the palette
sns.pairplot(df, diag_kind='kde', corner=True, markers='o')

# Add a title
plt.suptitle('Pair Plot : Age, Height, and Weight', y=1.02, fontsize=16, fontweight='bold')

# Show the plot
plt.show()
```

10. Save and Style Plots
‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
# Apply a built-in style
plt.style.use('ggplot')
plt.plot(x, y, label='cos(x)', color='purple', linewidth=2)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Plot of cos(x)')
plt.legend()
plt.savefig('line_plot.png', dpi=300)
plt.savefig('line_plot.pdf')
# Displaying
plt.show()
```
