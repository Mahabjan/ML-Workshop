# **Histograms for Key Features of Penguins**
**Culmen Length (culmen_length_mm):**
Likely shows a normal or slightly skewed distribution, indicating most penguins have similar beak lengths, with a few outliers.

**Culmen Depth (culmen_depth_mm):**
May have a bimodal distribution, suggesting different species or sexes with varying beak depths.

**Flipper Length (flipper_length_mm): **
Likely normally distributed, with most penguins having similar flipper lengths, although species differences may show slight variations.

**Body Mass (body_mass_g):**
Expected to show a bimodal distribution, reflecting the difference in body mass between species like Ad√©lie (lighter) and Gentoo (heavier).

```python
# Histogram for numerical features
sns.histplot(df['culmen_length_mm'])
plt.title('culmen Length Distribution')
plt.show()

sns.histplot(df['culmen_depth_mm'])
plt.title('culmen Depth Distribution')
plt.show()

sns.histplot(df['flipper_length_mm'])
plt.title('Flipper Length Distribution')
plt.show()

sns.histplot(df['body_mass_g'])
plt.title('Body Mass Distribution')
plt.show()
```

# Box Plots for  Key Features of Penguins
The box plots for culmen length, culmen depth, and flipper length provide insights into the distribution of these key physical traits in penguins. Culmen length shows a few outliers on the higher end, indicating some penguins with significantly longer beaks. Culmen depth has a consistent distribution with no major outliers, suggesting uniformity in this feature. Flipper length has a broad range with outliers on the higher side, suggesting some penguins possess notably larger flippers. Overall, the data indicates variation across these physical features, with flipper and culmen length showing more spread.

```python
# Box plot to check for outliers
sns.boxplot(x=df['culmen_length_mm'])
plt.title('culmen Length Box Plot')
plt.show()

sns.boxplot(x=df['culmen_depth_mm'])
plt.title('culmen Depth Box Plot')
plt.show()
sns.boxplot(x=df['flipper_length_mm'])
plt.title('flipper Length Box Plot')
plt.show()
```

# Bar Plots for Penguin Species and Sex Distribution
The bar plots show the distribution of penguins by species and sex. The species distribution plot reveals that the dataset has an almost equal representation of the three species (Adelie, Chinstrap, and Gentoo), with Gentoo being slightly more abundant. The sex distribution plot shows a near equal distribution of male and female penguins, indicating a balanced representation of both sexes in the dataset.




```python
# Bar plot for species distribution
sns.countplot(x='species', data=df)
plt.title('Penguin Species Distribution')
plt.show()

# Bar plot for sex distribution
sns.countplot(x='sex', data=df)
plt.title('Penguin Sex Distribution')
plt.show()
```

# Boxplot of Body Mass by Island
The boxplot shows the distribution of body mass for penguins across different islands (Torgersen, Biscoe, and Dream). The median body mass appears to be relatively similar across all three islands, but the range and spread of values vary. Torgersen Island has a more compact distribution with fewer outliers, while Biscoe and Dream Islands show a wider spread, particularly Dream Island, which has some extreme outliers on the higher side. This suggests that penguins on Dream Island may have more variation in body mass, including some much larger individuals.





```python
sns.boxplot(
    data=df,
    x="island",
    y="body_mass_g",
)
plt.title("Boxplot of Body Mass (g) by island")
plt.xlabel("Island")
plt.ylabel("Body Mass (g)")
plt.show()
```

# Body Mass Distribution by Species and Sex
The violin plot shows the body mass distribution of penguins by species (Adelie, Chinstrap, and Gentoo) and sex (male and female). The plot highlights the distribution of body mass within each species, with females generally having slightly lower body mass than males across all species. Gentoo penguins show the highest body mass overall, followed by Chinstrap and Adelie penguins. The width of the violins indicates the density of body mass values, with Adelie penguins having a narrower distribution, suggesting less variation in body mass, while Gentoo has a wider spread, indicating more variation.




```python
sns.violinplot(
    data=df,
    x="species",
    y="body_mass_g",
    hue="sex",
    split=True,
)
plt.title("Body Mass Distribution by Species and Sex")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.legend(title="Sex")
plt.show()
```

# Correlation Matrix of Penguin Features
The heatmap of the correlation matrix illustrates the relationships between key physical features of penguins: culmen length, culmen depth, flipper length, and body mass. The heatmap shows that culmen length and culmen depth are moderately positively correlated, meaning that penguins with longer beaks tend to have greater beak depth. There is also a strong positive correlation between flipper length and body mass, indicating that larger penguins tend to have longer flippers. However, the correlations between other pairs of features, such as culmen length and flipper length, are relatively weaker, suggesting that these features vary independently to some extent.


```python
correlation_matrix = df[["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")
plt.title("Correlation Matrix of Penguin Features")
plt.show()
```

# Pairplot of Penguin Features
The pairplot visualizes the relationships between multiple physical features of penguins, with points colored according to species (Adelie, Chinstrap, and Gentoo). The diagonal shows the distribution of each feature, while the off-diagonal plots illustrate the pairwise relationships between them. There are noticeable separations between species in certain pairs, particularly in features like body mass and flipper length, suggesting that these species differ in their physical characteristics. Gentoo penguins tend to have higher values for both body mass and flipper length, while Adelie and Chinstrap penguins have more overlap.

```python
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Penguin Features')
plt.show()
```

# Word Cloud of Species Names
The word cloud visualizes the frequency of each species name in the dataset, with the size of each word representing how frequently it appears. The larger words indicate that the corresponding species appears more often in the dataset, while smaller words suggest lower frequency. This provides a quick, visual representation of the species distribution, making it clear which species are more prevalent in the dataset.




```python
from wordcloud import WordCloud
species_names = df['species'].tolist()
text = ' '.join(species_names)
wordcloud = WordCloud(width=600, height=400, background_color='pink').generate(text)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bicubic')
plt.axis('off')
plt.title("Word Cloud of species names")
plt.show()
```
